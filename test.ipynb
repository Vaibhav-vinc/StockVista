## Check age of saved files

```python
import os, time
import datetime as dt
```

```python
file = time.ctime(os.path.getctime('./media/AAPL.csv'))
```

```python
file = file.split()[3]
```

```python
file_hour = int(file.split(':')[0])
file_minute = int(file.split(':')[1])
file_seconds = int(file.split(':')[2])
```

```python
filetime = dt.timedelta(hours=file_hour, minutes=file_minute, seconds=file_seconds)
```

```python
current = dt.datetime.now()
```

```python
current = str(current).split()[1].split('.')[0]
```

```python
current_hour = int(current.split(':')[0])
current_minute = int(current.split(':')[1])
current_seconds = int(current.split(':')[2])
```

```python
currenttime = dt.timedelta(hours=current_hour, minutes=current_minute, seconds=current_seconds)
```

```python
filetime
```

```python
currenttime
```

```python
diff = (currenttime - filetime)
min = diff / 60
```

```python
diff
```

```python
min.seconds >= 10
```
